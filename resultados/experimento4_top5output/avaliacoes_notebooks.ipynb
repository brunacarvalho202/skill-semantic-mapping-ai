```python
import pandas as pd

df_final = pd.read_csv('mapeamento_final_bi_cross.csv')
df_audit = pd.read_csv('auditoria_necessaria.csv')

total = len(df_final)
matches = df_final[df_final['match_status'] == 'matched']
no_matches = df_final[df_final['match_status'] == 'no_match']

print(f"Total de registros: {total}")
print(f"Matches confirmados: {len(matches)} ({len(matches)/total:.1%})")
print(f"No Matches (Gaps): {len(no_matches)} ({len(no_matches)/total:.1%})")
```

```python
#filtre df_final para ter apenas as linhas de ai_rerank na coluna strategy_used
df_audit = df_final[df_final['strategy_used'] == 'ai_rerank']
df_audit.head(1)
```

```python
#media de diferenca entre o score_bi_encoder_top1 e o score_bi_encoder_top2
df_audit['score_diff'] = df_audit['score_bi_encoder_top1'] - df_audit['score_bi_encoder_top2']
print(f"Média da diferença entre top1 e top2: {df_audit['score_diff'].mean():.4f}") 
```

- A média de diferença entre top1 e top2 me mostra uma baixa margem de confiança ou uma alta ambiguidade entre os candidatos

Com essa diferença pequena, qualquer ruído nos dados pode inverter a ordem, o que torna a solução instável.
A solução apresenta baixa distinção semântica entre as melhores opções.

- Oções de próximos experimentos para melhorar essa confiança:

1. Enriquecimento de contexto das skills, manualmente preparados, por LLM ou RAG para maior controle de alucinação.

2. Reestruturar a taxonomia atual:  a taxonomia atual está muito simples e pode deixar o modelo confuso com skills parecidas ou que são categorias e subcategorias.

3. Melhorar modelo do Cross-Encoder: experimentar se um modelo de reranking mais robusto ou ajustes na temperatura da predição ajudariam a aumentar essa margem.
